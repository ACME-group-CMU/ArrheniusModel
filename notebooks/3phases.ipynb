```julia
# Initial composition
existing_layers = [[1.0, .0, .0]]

# Composition of new layer to deposit at each time step
new_layer = [1.0, .0, .0]

# Example rate coefficients and activation energies for each phase
A_values = [1.0, 1.0, 1.0]
Ea_values = [50000, 60000, 70000]
T = 300  # Kelvin

# # of time steps
num_steps = 10

# Simulate deposition and transformation
for step in 1:num_steps
    existing_layers = arrhenius_transform_with_deposition(existing_layers, new_layer, A_values, Ea_values, T)
    println("Step $step: ", existing_layers)
end
```
